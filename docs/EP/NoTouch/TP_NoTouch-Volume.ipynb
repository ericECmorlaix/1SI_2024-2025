# NoTouch

## Prsentation

NoTouch est un distributeur automatique de gel lavant (savon liquide).



Dans un souci de sant publique, il est ncessaire de limiter la transmission des bactries lors du lavage des mains.  
Le problme vient du fait qu'avec les distributeurs de savon manuels, le contact des mains favorise la transmission des bactries.

Le systme tudi permet la distribution sans contact d'une dose suffisante de savon liquide pour un usage domestique.
