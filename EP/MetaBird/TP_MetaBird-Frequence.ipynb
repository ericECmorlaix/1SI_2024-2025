# MetaBird

## Prsentation

L'oiseau connect Metabird cr par la socit [BionicBird](https://www.bionicbird.com/fr/metabird/) est un drone pilotable par smartphone qui vole en battant des ailes.



La conception de ce drone se base sur le principe de biomimtisme. Il sinspire et reproduit de faon raliste le vol dun oiseau.   
Il est un concentr de technologie: ailes souples en mylar et fibre de carbone, un corps en mousse indestructible, un poids de moins de 10g,
Batterie lithium-ion polymre Li-Po de 3,7V et une carte [SoC](https://fr.wikipedia.org/wiki/Systme_sur_une_puce) miniature.  
Connect en Bluetooth, il se pilote  laide dune application sur un smartphone, utilisant les acclromtres et gyroscopes du tlphone pour le diriger.  
Labsence dhlice et ses lments de scurit font galement de loiseau connect un objet inoffensif pour ses utilisateurs.  
Il permet des vols denviron 7 minutes et se recharge en seulement 13 minutes.
